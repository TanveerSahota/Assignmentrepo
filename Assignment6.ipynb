```python

print("Ques 1.")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sls
import pandas as pd

M = int(input("Enter value for M: "))

x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8,6))
plt.plot(x, y1, label=f'y = {M}x^2', linestyle='-', color='g')
plt.plot(x, y2, label=f'y = {M}sin(x)', linestyle='--', color='r')
plt.legend()
plt.grid()
plt.title("Graph Plots")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
print("Ques 2.")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sls
import pandas as pd

sc=np.random.randint(50,100,5)
data = {
    'Subjects': ['Math', 'sst', 'english', 'punjabi', 'art'],
    'Scores': sc
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,6))
ax = sls.barplot(x='Subjects', y='Scores', data=df, palette='viridis')

for p in ax.patches:
    ax.annotate(f"{p.get_height()}", (p.get_x() + p.get_width()/2, p.get_height()),
                 ha='center', va='bottom', fontsize=12)

plt.title("Scores of 5 subjects")
plt.xlabel("subjects")
plt.ylabel("scores")
plt.grid(axis='y')
plt.show()
```

```python
print("Ques 3.")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sls
import pandas as pd

rollnumber = 102317194
np.random.seed(rollnumber)

data = np.random.randn(50)

cumsum = np.cumsum(data)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].plot(cumsum,marker='o', color='g', label="cumulative Sum")

axes[0].set_title("cumulative sum")
axes[0].legend()
axes[0].grid()

axes[1].scatter(range(50), data, c='red', label="random noise")
axes[1].set_title("Scatter Plot with Noise")
axes[1].legend()
axes[1].grid()

plt.tight_layout()
plt.show()
```

```python
print("Ques 4.")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)

plt.figure(figsize=(8,6))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='c')
plt.title("total profit over months ")
plt.xlabel("month")
plt.ylabel("Total profit ")
plt.grid()
plt.show()

plt.figure(figsize=(8,6))
for column in df.columns[1:-1]:
    sns.lineplot(x=df['month_number'], y=df[column], label=column)
plt.title("product sales over months ")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.legend()
plt.grid()
plt.show()

df.plot(kind='bar', figsize=(12,6), alpha=0.75)
plt.title("Sales Data")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend(loc='upper right')
plt.show()
```
